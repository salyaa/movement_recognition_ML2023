# Pre-processing of our data and creation of our machine learning entities X and y:

# Separation of Train-Test using the first method:

The first method consist of removing the time column from `X` and a split of 50% for
testing and 50% for training is performed randomly.

```python
X_names_M1 = list(('X_train_M1', 'X_test_M1'))

test_labels_M1, accuracy_M1 = train_test_clustering_svm(X, X_names_M1, y_encoded, 'M1', True, True)
```

We can now start our classification for each matrix and compare our result.

```python
# Frontal Top
X_names_ft_M1 = list(('X_Frontal_Top_train_M1', 'X_Frontal_Top_test_M1'))
test_labels_ft_M1, accuracy_ft_M1 = train_test_clustering_svm(X_ft, X_names_ft_M1, y_encoded_ft, 'M1', True, False)

# Frontal Low
X_names_fl_M1 = list(('X_Frontal_Low_train_M1', 'X_Frontal_Low_test_M1'))
test_labels_fl_M1, accuracy_fl_M1 = train_test_clustering_svm(X_fl, X_names_fl_M1, y_encoded_fl, 'M1', True, False)

# Side Top
X_names_st_M1 = list(('X_Side_Top_train_M1', 'X_Side_Top_test_M1'))
test_labels_st_M1, accuracy_st_M1 = train_test_clustering_svm(X_st, X_names_st_M1, y_encoded_st, 'M1', True, False)

# Side Low
X_names_sl_M1 = list(('X_Side_Low_train_M1', 'X_Side_Low_test_M1'))
test_labels_sl_M1, accuracy_sl_M1 = train_test_clustering_svm(X_sl, X_names_sl_M1, y_encoded_sl, 'M1', True, False)
```

```python
# Let us compute the average accuracy when looking at the different cameras:
accuracy_cameras_M1 = np.array([accuracy_ft_M1, accuracy_fl_M1, accuracy_st_M1, accuracy_sl_M1])
avg_accuracy_M1 = accuracy_cameras_M1.mean()
print(f'Average accuracy: {100*avg_accuracy_M1}%')
```

```python
# Now, let us compute the difference in accuracy between our two different results:
difference_in_accuracy_M1 = (avg_accuracy_M1 - accuracy_M1)
print(f'The difference in accuracy is: {difference_in_accuracy_M1*100}%')
```

# What axis is the most important one?

Since the difference in accuracy is almost always the same for each splitting, we will look at the whole matrix rather than the sub-matrices for each camera and use the third splitting method, since the results will be similar.

```python
remove_columns_x = [col for col in X.columns if col[-1]=='x']
X_no_x = X.drop(remove_columns_x, axis='columns')

X_names_x = list(('X_train_no_x', 'X_test_no_x'))

test_labels_x, accuracy_x = train_test_clustering_rf(X_no_x, X_names_x, y_encoded, 'M3', False, False)
```

```python
remove_columns_y = [col for col in X.columns if col[-1]=='y']
X_no_y = X.drop(remove_columns_y, axis='columns')

X_names_y = list(('X_train_no_y', 'X_test_no_y'))

test_labels_y, accuracy_y = train_test_clustering_rf(X_no_y, X_names_y, y_encoded, 'M3', False, False)
```

```python
remove_columns_z = [col for col in X.columns if col[-1]=='z']
X_no_z = X.drop(remove_columns_z, axis='columns')

X_names_z = list(('X_train_no_z', 'X_test_no_z'))

test_labels_z, accuracy_z = train_test_clustering_rf(X, X_names_z, y_encoded, 'M3', False, False)
```

```python
# For the x-axis:
difference_x = accuracy_M3 - accuracy_x
print(f'The difference in accuracy for removing the x-axis is: {difference_x*100:.4f}%.')
# For the y-axis:
difference_y = accuracy_M3 - accuracy_y
print(f'The difference in accuracy for removing the y-axis is: {difference_y*100:.4f}%.')
# For the z-axis:
difference_z = accuracy_M3 - accuracy_z
print(f'The difference in accuracy for removing the z-axis is: {difference_z*100:.4f}%.')
```

Hence, removing the x-axis has the most impact with this splitting method.
